````python
import random
import time
names=["pratik","rika","papali","pupuli","puja","Rohith","priya"]
subjects=["django","flask","heroku","postgres","git"]
# num=int(input("Enter the Number of student list"))
def student_list(num):
#     students=[]
    for x in range(num):
        student={"id":x,"name":random.choice(names),"subj":random.choice(subjects)}
        time.sleep(1)
        yield student
#         time.sleep(1)
    return student

g1=student_list(100)
t1=time.time()
for x in g1:
    print(x)
t2=time.time()        
print(g1)
print(t2-t1)
````

````python
#here we want to checking how to convert the generator object to list
#there are 2 ways are there
#1st way:-
def natural_generator(n):
    i=1
    while i<=n:
        yield i
        i+=1
        n-=1
# it=natural_generator(10)
# print(it)
# l1=list(it)
# print(l1)

#2nd way:-
it=natural_generator(10)
l2=[]
while True:
    try:
        l2.append(next(it))
    except StopIteration:
        break
print(l2)
````

````python
def range_gen(start,end):
    while start<=end:
        yield start
        start=start+1

my_range_it=range_gen(10,20)
while True:
    try:
        print(next(my_range_it))
    except StopIteration:
        break
    
````
