````python
#iterating each link then
r=requests.get("https://formulae.brew.sh/api/formula.json")
resp=r.json()
output=[]
time_started=time.perf_counter
for package in resp:
    package_name=package.get("name")
    package_desc=package["desc"]
    package_url=f"https://formulae.brew.sh/api/formula/{package_name}.json"
    response1=requests.get(package_url).json()
    install_30=response1["analytics"]["install"]["30d"][package_name]
    install_90=response1["analytics"]["install"]["90d"][package_name]
    install_365=response1["analytics"]["install"]["365d"][package_name]
#     print(package_name,package_desc,install_30,install_90,install_365)
    print(f"package{package_name} finished in {requests.get(package_url).elapsed} seconds ")
    data=[{
        "name":package_name,
        "desc":package_desc,
        "analytics":{"30d":install_30,"90d":install_90,"365d":install_365}
    }]
    output.append(data)
    time.sleep(requests.get(package_url).elapsed.total_seconds())
# print(output)
time_ended=time.perf_counter
# print(f"total time is {time_ended-time_started}")
with open ("package.json","w") as f1:
    json.dump(output,f,indent=2)
    
    
````
